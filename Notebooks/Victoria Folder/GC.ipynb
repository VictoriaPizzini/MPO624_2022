```python
import numpy as np
import matplotlib.pyplot as plt

def p_L(z):
    return 1010*((305-4*z)/305)**(9.81/(287*4))

def p_0(z):
    return 1015*((300-6*z)/300)**(9.81/(287*6))

z = np.linspace(0,10,1000)

plt.plot(z,p_L(z), color='orange')
plt.plot(z,p_0(z), color='green')
plt.xlabel("z (km)")
plt.xlabel("P (Pa)")

print(p_0(z)-p_L(z))
```

```python
plt.plot(z, p_0(z) - p_L(z), color='red')
plt.xlabel("z (km)")
plt.ylabel("P (Pa)")
```
